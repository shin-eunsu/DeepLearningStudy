## yolo labeling 파일명 = class명
## 파일명에 맞춰 class 정보 수정

```python
import os
import shutil
```

```python
txtDirPath = 'D:/DataSet/image/'
fileList = os.listdir(txtDirPath)
txtList = [file for file in fileList if file.endswith(".txt")]
```

```python
classList = []
for txt in txtList:
    txtPath = txtDirPath + txt
    
    if txt == "classes.txt":
        with open(txtPath, 'r') as f:
            for line in f.readlines():
                classList.append(line.strip('\n'))               
    else:
        newContents = ''
        productName = txt.split('_')[0]
        lineCnt = 0
        with open(txtPath) as f:
            for line in f.readlines():
                newContents += line
                lineCnt += 1
        txtLine = newContents.split('\n')
        
        for i in range(len(txtLine) - 1):
            txtLine[i] =txtLine[i].split(' ')
            txtLine[i][0] = classList.index(productName)
    
        for i in range(len(txtLine)):
#             print(txtLine)
            for j in range(len(txtLine[i])):
                print(txtLine[i][j])
                txtLine[i][j] = str(txtLine[i][j])                
        
        saveContents = ''
        for i in range(len(txtLine) - 1):
            saveContents += ' '.join(txtLine[i]) + '\n'
#         print(saveContents)
        
        with open(txtPath, 'w') as f:
            f.write(saveContents)
            
```
